Introduction:

As a Southern California native turned expatriate, I observed the cultural melting pot of Los Angeles county at various points throughout my childhood and in varying ways. I've seen racism perpetrated by people of every color, sometimes personally and sometimes systemically without any intention at all. However, I also left California when I became an adult, and so missed out on several years worth of change in the California social, political and even legal landscape between 2006 and 2020. It was then, with some interest, that I came across a dataset published for compliance with the FBI's 'National Incident-Based Reporting System'. This data detailed approximately 1mln crime incidents, including the date of report, date and time the incident occurred, the Los Angeles Police Division with jurisdiction over the incident and precise coordinates for the incident itself, various categorizations of incidents by both a crime code given by the LAPD and an incident description, victim characterization (though notably no characterization of a suspect), and a status on the outcome of the incident itself.

This dataset provides a great deal of useful detail, even if there are some theoretical issues with it. The data gives us the flexibility to answer a few questions with regard to changes that the COVID-19 pandemic had on the nature of crimes in Los Angeles, as well as provide metrics on how effective the LAPD was at adjusting to the challenges posed by COVID-19. We aim to test the hypothesis that the pandemic -- and the social distancing that was required by it -- caused crimes to change in nature by either intention or circumstance into types that are inherently resistant to policing, and further did nothing to prevent disparity in treatment across racial and ethnic groups.

First, we will discuss the limitations of the dataset, and then review the state of the environment at the start of the data collection -- elements that might have a confounding impact on the data we discover. Then we will seek to answer three questions that serve to help inform our hypothesis: what crimes comprised the time period covered in the dataset and how is it different from before the dataset collection began, are their arrest rate disparities in the demographics of the victims, and finally observe the changes to Intimate Partner Violence during the reported period to see if COVID-19 and resulting changes in daily life had impacts on it, and how those impacts differ from other types of assault in the reported period. Combining these three sets of information may shed some light on the changing nature of law enforcement and social dynamics during the pandemic and subsequent reintegration. Based on our findings, we may may raise questions worth considering for the future of law enforcement policy and practice.

In 2014, California passed Proposition 47, which changed specific property and drug charges from felony to misdemeanor. This led to a decrease in racial disparity in arrests and bookings, though fissures still remained -- specifically with respect to arrest rates against African Americans. (Cite - Goldrush Cam) In a broader sense, California's criminal justice reform had been largely successful in June of 2019, with the incarceration rate dropping from 476 to 314 per 100,000 residents compared to the same month in 2006.(cite - California Budget and Policy Center, Goldrush Cam) This suggests that criminal reform to this point had been largely successful in LA County. In short -- California law reforms in the 2010s had largely been effective at reducing reported criminal activity.

With regard to events during the reported period that may be of note as markers for activity, there were three notable reforms passed during the first year of the data -- 2020. These include AB 2542, AB 1950, and SB 823. AB 2542, or the California Racial Justice act, prevented prosecutors from imposing a conviction or sentence based purely on race and allowed defendence to issue challenges in that regard.(cite ABA) AB 1950 adjusted maximum probationary periods for felonies and misdemeanors, with the former decreasing from five years to two, and the later from three years to one.(cite BrownsteinLawGroup) SB 823 closed the state 'Division of Juvinile Justice', which gave county justice systems the ability to assist adolescents in need of guidance rather than hamstring their chance at a future.(cite BrownsteinLawGroup) The overall theme of this changing legislation is one of growth towards racial equity -- by preventing racially motivated prosecution it became harder for prejudiced prosecutors or judges to advance their beliefs. By reducing probationary periods for felonies and misdemeanors, focus was taken off of wasteful practicies that did little to reduce recidivism while encouraging greater legal exposure that would invite further technical violations. Finally, by offering youth the opportunity to change their trajectory back into something more socially acceptible, less repeat offenses were encouraged by the judicial system. This is the scene that our dataset opens on -- a county that has made significan strides towards preventing inequity, and has started to see dividends on the investment in the form of a lower reported crime rate.
